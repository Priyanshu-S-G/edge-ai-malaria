```python
import os
import sys
import numpy as np
import kagglehub
```

```python
path = kagglehub.dataset_download("iarunava/cell-images-for-detecting-malaria")

RAW_DIR = os.path.join(path, "cell_images", "cell_images")

print(os.listdir(RAW_DIR))
```

```python
SRC_PATH = "/kaggle/input/datasets/priyanshusghosh/source/src"
sys.path.append(SRC_PATH)

from data.loader import load_image_paths
```

```python
data = load_image_paths(RAW_DIR)

print("Total samples:", len(data))
```

```python
indices = np.arange(len(data))
```

```python
labels = np.array([label for _, label in data])

# Separate indices by class
idx_class0 = indices[labels == 0]
idx_class1 = indices[labels == 1]

# Shuffle deterministically
np.random.seed(42)
np.random.shuffle(idx_class0)
np.random.shuffle(idx_class1)

# 80-20 split
split0 = int(0.8 * len(idx_class0))
split1 = int(0.8 * len(idx_class1))

train_idx = np.concatenate([idx_class0[:split0], idx_class1[:split1]])
test_idx  = np.concatenate([idx_class0[split0:], idx_class1[split1:]])

# Shuffle final sets
np.random.shuffle(train_idx)
np.random.shuffle(test_idx)

print("Train:", len(train_idx))
print("Test:", len(test_idx))
```

```python
SPLIT_DIR = "/kaggle/working/splits"
os.makedirs(SPLIT_DIR, exist_ok=True)

np.save(os.path.join(SPLIT_DIR, "train_idx.npy"), train_idx)
np.save(os.path.join(SPLIT_DIR, "test_idx.npy"), test_idx)
```

```python
def check_balance(idx, name):
    cls = labels[idx]
    unique, counts = np.unique(cls, return_counts=True)
    print(name, dict(zip(unique, counts)))

check_balance(train_idx, "Train")
check_balance(test_idx, "Test")
```
